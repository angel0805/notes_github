#### **Example: Tossing Coins**

Consider the example of tossing a fair coin to root our vocabulary and symbols. We will say the probability of getting a head is the total ways for this event to happen divided by the number of all possible outcomes

**Example I:** A Fair Coin

With one fair coin, we have outcomes wither heads or tails. The outcome of interest is either a heads or tails. We would say the probability of each is $\frac{1}{2}$

$$P(\text{head}) = \frac{1}{2} \quad P(\text{tails}) = \frac{1}{2}$$

Note that these probabilities sum to 1.

**Example II:** An Unfair Coin

An example of an unfair coin would be anything that does not have equal probability. For example:

$$P(\text{head}) = \frac{6}{10} \quad P(\text{tails}) = \frac{4}{10}$$

```python
import scipy.stats as stats

fig, ax = plt.subplots(1, 2, figsize = (20, 5))

fair_coin = stats.bernoulli(.5).pmf([0, 1])
ax[0].bar(['heads', 'tails'], fair_coin, color = ['gray', 'white'], edgecolor = 'black')
ax[0].set_title('Probabilities for a fair coin')

unfair_coin = stats.bernoulli(.4).pmf([0, 1])
ax[1].bar(['heads', 'tails'], unfair_coin, color = ['gray', 'white'], edgecolor = 'black')
ax[1].set_title('Probabilities for an unfair_coin')

plt.show()
```

#### Probability Mass Function

We can describe this kind of events as a function. Specifically, a situation in which there are two possible outcomes with a probability attached to each. This is called a *Bernoulli* distribution, and the functional form is:

$$f(k;p) = pk + (1 - p)(1 - k) \quad {\text{for }}k\in \{1, 0\}$$

In our example of an unfair coin, if we consider heads as 0 and tails as 1 ($k$), we would have:

$$ f(0, .4) = .4 * 0 + (1 - .4)(1 - 0) $$

which equals 0.6. Similarly, the probability of a tails ($k = 0$)

$$ f(1, .4) = .4 * 1 + (1 - 1)(1 - 1) $$

Which equals 0.4

```python
def bernoulli(k, p):
  return p * k + (1 - p) * (1 - k)
```

```python
bernoulli(0, 0.4)
```

```python
bernoulli(1, 0.4)
```

1. Suppose a bag contains 8 Red Marbles and 3 Blue Marbles.
 - What is the probability of choosing a red ball from a random draw?
 - And the probability of a blue ball?

```python
import matplotlib.pyplot as plt
from scipy.stats import binom

p_red = 8/11
p_blue = 3/11

# What is the probability of choosing a red ball from a random draw?
pmf_red = binom.pmf(1, 1, p_red)
print(f"Answer: {pmf_red:.2f}%")

# And the probability of a blue ball
pmf_blue = binom.pmf(1, 1, p_blue)
print(f"Answer: {pmf_blue:.2f}%")

fig, ax = plt.subplots(1, 2, figsize = (10, 5))

ax[0].stem(1, pmf_red, 'r-')
ax[0].set_title('Probabilities of red ball')

ax[1].stem(1, pmf_blue, 'b-')
ax[1].set_title('Probabilities of blue ball')

plt.show()
```

2. Suppose a bag contain 4 Red Marbles, 7 Yellow Marbles, and 2 Blue Marbles.
 - What is the probability of selecting a red or yellow marble?
 - What is the probability of selecting a yellow or blue marble?
 - How many ways can you select two colors of marbles from the bag?

```python
import math
from scipy.special import comb

# Total de canicas
total_marbles = 13

# Cantidad de cada canica
red_marbles = 4
y_marbles = 7
blue_marbles = 2

# Obtenemos las probabilidades de cada canica
p_r_marbles = red_marbles / total_marbles # probabilidad de seleccionar una canica roja
p_y_marbles = y_marbles / total_marbles # probabilidad de seleccionar una canica amarilla
p_b_marbles = blue_marbles / total_marbles # probabilidad de seleccionar una canica azul+

# ##### Task 1 #####
# Sumamos las probabilidades de red marbles y yellow marbles
p_r_y_marbles = p_r_marbles + p_y_marbles
print(f"The probability of selecting a red or yellow marbles is : {(p_r_y_marbles)*100:.2f}%")

# ##### Task 2 #####
# Sumamos las probabilidades de yellow marbles y blue marbles
p_y_b_marbles = p_y_marbles + p_b_marbles
print(f'The probability of selecting a yello or blue marbles is : {(p_y_b_marbles)*100:.2f}%')
```

Usamos la formula de combinacion para calcular el numero de formas de seleccionar dos colores de canicas

$$ C(n,k) = \frac{n!}{k!(n-k)!} $$

```python
# ##### Task 3 #####

# Creamos una funcion que calculara la cantidad de combinaciones de seleccionar 2 colores de canica en la bolsa
def calcular_combinaciones(total_elementos, cantidad_seleccionar):
  return math.factorial(total_elementos) / (math.factorial(cantidad_seleccionar) * math.factorial(total_elementos - cantidad_seleccionar))

print(f"Way_1 - Number of ways to select two color in marbles bag: {calcular_combinaciones(total_marbles, 2)}")

# Another way
print(f"Way_2 - Number of ways to select two color in marbles bag: {comb(total_marbles, 2)}")
```

3. Suppose a bag contains 10 Red and 7 Blue marbles.
 - Suppose you select one marble from the bag, what is the probability this is red?
 - Suppose you select a blue marble, place the marble back in the bag and select another marble. What is the probability the second marble is blue?
 - Suppose you select a blue marble, do not replace the marble the bag and select another marble. What is the probability the second marble is blue?

```python
# Total canicas
total_marbles = 17

# Cantidad de cada tipo de canica
red_marbles = 10
blue_marbles = 7

# Probabilidad por color
p_r_marble = red_marbles / total_marbles # Canicas rojas
p_b_marble = blue_marbles / total_marbles # Canicas azules

# Tarea 1: Probabilidad de sacar una canica roja
print(f"La probabilidad de sacar una canica que sea roja es: {p_r_marble*100:.2f}%")

# Tarea 2: Probabilidad de que la segunda canica sea azul (con reemplazo)
# Como la canica se vuelve a poner en la bolsa, la cantidad de canicas se mantiene y la probabilidad no cambia
print(f"La probabilidad de que la segunda canica sea azul es: {p_b_marble*100:.2f}%")

# Tarea 3: Probabilidad de que la segunda canica sea azul (sin reemplazo)
# Se saca una canica azul y no se vuelve a poner en la bolsa
p_b_replacement = (blue_marbles - 1) / (total_marbles - 1)
print(f"La probabilidad de sacar una canica azul despues de haber sacado una canica azul es: {p_b_replacement*100:.2f}%")

```
