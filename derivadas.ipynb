```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sy
```

Consider:

$$
 f(x) = x + 1 \ \ and \ \ g(x) = x^2 + 1
$$

```python
def f(x): return x + 1 # Definimos una funcion `f(x)` que toma `x` como parametro y devuelve `x + 1`
def g(x): return x**2 + 1 # Definimos una funcion `g(x)` que toma `x` como parametro y devuelve `x` al cuadrado + 1
x = np.linspace(start = -2, stop = 2, num = 100) # Esto crea un array de 100 numero igualmente espaciados entre `-2` y `2`
```

```python
plt.plot(x, f(x), label = 'f') # Traza la funcion `f(x)` en el grafico con `x` en el eje x y `f(x)` en el eje y.
plt.plot(x, g(x), label = 'g') # Traza la funcion `f(x)` en el grafico con `x` en el ejex y `f(x)` en el eje y.
plt.plot(0, f(0), 'ro') # Traza un punto rojo en la coordenada (0, f(0))
plt.plot(1, f(1), 'ro') # Traza un punto rojo en la coordenada (1, f(1))
plt.legend()
```

```python
x2 = np.linspace(start = 1, stop = 1.1)
```

```python
plt.plot(x2, f(x2), label = 'f')
plt.plot(x2, g(x2), label = 'g')
plt.legend()
plt.title('Zooming in on $x = [1, 1.1]$\nNow they both look Linear!')
```

Definition

$$
f'(x) = \lim_{h \to 0} \frac{f(x + h) - f(x)}{h} \
$$

## Derivative as a Function with Python

```python
def f(x):
  return x**2 - 2*x + 1
```

```python
def df(x):
  h = 0.000001
  return (f(x + h) - f(x))/h
```

```python
x = np.linspace(-4, 4, 1000)
plt.plot(x, f(x), label = '$f(x)$')
plt.plot(x, df(x), label = '$f\'(x)$')
plt.axhline(color = 'black')
plt.axvline(color = 'black')
plt.legend(loc = 'best', frameon = False)
plt.title('Function and its Derivative')
```
